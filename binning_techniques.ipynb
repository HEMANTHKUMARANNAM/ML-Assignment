```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
data = [
    [1,15,70,'maths'],
    [2,19,60,'maths'],
    [3,20,60,'science'],
    [4,15,75,'maths'],
    [5,np.nan,78,'maths'],
    [6,18,np.nan,'science'],
    [7,10,80,'maths']
]

cols = ['Id', 'Age' , 'Marks' , 'Name']
```

```python
df = pd.DataFrame(data, columns = cols)
df
```

```python
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df
```

```python
df_sorted_by_age = df.sort_values('Age', na_position='last')
df_sorted_by_age
```
