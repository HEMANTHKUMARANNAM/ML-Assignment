```python
# Function for log transformation of the column
def log_transform(data3,col):
    for colname in col:
          if (data3[colname] == 1.0).all():
              data3[colname + '_log'] = np.log(data3[colname]+1)
          else:
              data3[colname + '_log'] = np.log(data3[colname])
```

```python
log_transform(marutidata,['TotalSteps','TotalDistance'])
```
